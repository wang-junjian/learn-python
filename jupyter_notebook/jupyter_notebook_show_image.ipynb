## 解决颜色显示不正确的问题
* [OpenCV与Python之图像的读入与显示以及利用Numpy的图像转换](https://www.cnblogs.com/visionfeng/p/6094423.html)
* [Python图像处理（2）：图像显示](https://blog.csdn.net/lights_joy/article/details/45933907)

```python
img = cv2.imread('test.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img)
plt.show()
```

## 移除X轴和Y轴的tick
* [Remove xticks in a matplotlib plot?](https://stackoverflow.com/questions/12998430/remove-xticks-in-a-matplotlib-plot)

```python
img = cv2.imread('test.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.xticks([])
plt.yticks([])
plt.imshow(img)
plt.show()
```

## 在Jupyter Notebook里显示多张图片
* [python使用matplotlib:subplot绘制多个子图](https://blog.csdn.net/gatieme/article/details/61416645)

```python
img_bgr = cv2.imread('test.jpg')
img_rgb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)

plt.figure(figsize=(10, 4))

# 1行2列第1个格
plt.subplot(121)
plt.title('OpenCV BGR')
plt.imshow(img_bgr)

# 1行2列第2个格
plt.subplot(122)
plt.title('Matplot RGB')
plt.imshow(img_rgb)

plt.show()
```
