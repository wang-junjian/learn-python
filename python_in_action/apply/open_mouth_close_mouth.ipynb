## facial landmarks 68 markup

[Dlib facial landmark starting index](https://stackoverflow.com/questions/41794191/dlib-facial-landmark-starting-index)

```python
import os
import matplotlib.pyplot as plt
import face_recognition

from PIL import Image
from PIL import ImageDraw
```

```python
"""显示所有的测试图片"""
def show_images(imgs):
    plt.figure(figsize=(20, 6))

    for i, img in enumerate(imgs):
        plt.subplot(1, len(imgs), i+1)
        plt.imshow(img)

    plt.show()
    
"""加载图片数据"""
def load_images(path):
    imgs = []
    for filename in os.listdir(path):
        if filename == '.DS_Store':
            continue

        file_path = os.path.join(path, filename)
        imgs.append(face_recognition.load_image_file(file_path))
    
    return imgs
    
"""加载图片路径"""
def load_image_paths(path):
    paths = []
    for filename in os.listdir(path):
        if filename == '.DS_Store':
            continue

        file_path = os.path.join(path, filename)
        paths.append(file_path)
    
    return paths
```

```python
"""画出嘴的轮廓"""
def draw_mouth(open_mouth_image_paths):
    imgs = []
    for img_path in open_mouth_image_paths:
        landmark = face_recognition.face_landmarks(face_recognition.load_image_file(img_path))[0]
        mouth_landmark = landmark['top_lip'][6:]+landmark['bottom_lip'][6:]

        img = Image.open(img_path)
        draw = ImageDraw.Draw(img)
        draw.polygon(mouth_landmark, fill=(255, 0, 0, 255))

        imgs.append(img)

    show_images(imgs)
```

```python
open_mouth_image_path = 'images/open_mouth_close_mouth/open_mouth/'
close_mouth_image_path = 'images/open_mouth_close_mouth/close_mouth/'
```

## 张嘴

```python
open_mouth_images = load_images(open_mouth_image_path)
show_images(open_mouth_images)
```

```python
open_mouth_image_paths = load_image_paths(open_mouth_image_path)
draw_mouth(open_mouth_image_paths[:4])
draw_mouth(open_mouth_image_paths[4:])
```

## 闭嘴

```python
close_mouth_images = load_images(close_mouth_image_path)
show_images(close_mouth_images)
```

```python
close_mouth_image_paths = load_image_paths(close_mouth_image_path)
draw_mouth(close_mouth_image_paths[:4])
draw_mouth(close_mouth_image_paths[4:])
```
